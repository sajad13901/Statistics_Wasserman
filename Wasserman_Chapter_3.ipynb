```
# Chapter 3, problem 9
import numpy as np
import matplotlib.pyplot as plt

N = 1000
vec_n = np.linspace(1, N, num=N)
tmp = np.random.normal(0, 1, N)

vec_xb = []
for i in vec_n:
  tmp = np.random.normal(0, 1, int(i))
  vec_xb.append(sum(tmp)/len(tmp))
plt.scatter(vec_n, vec_xb, label='Normal dist')
plt.legend(loc='upper right')
plt.show()

vec_yb = []
for i in vec_n:
  tmp = np.random.standard_cauchy(int(i))
  vec_yb.append(sum(tmp)/len(tmp))
plt.scatter(vec_n, vec_yb, label='Cauchy dist')
plt.legend(loc='upper right')
plt.show()
```

```
# Chapter 3, problem 10
import numpy as np
import matplotlib.pyplot as plt

N = 1000000
x = np.random.normal(0, 1, N)
y = np.exp(x)
y2 = pow(y, 2)

print('\t expectation of y')
print(sum(y)/len(y))
print(pow(np.exp(1), .5) )

print('\t expectation of y^2')
print(sum(y2)/len(y2))
print(pow(np.exp(1), 2) )

print('\t variance of y')
print(np.std(y) ** 2)
print(np.exp(1)**2 - np.exp(1) )
```

```
# Chapter 3, problem 11
import numpy as np

N = 10000
vec_n = np.linspace(1, N, N)

vec_x = []
for n in vec_n:
  y = np.random.choice([1, -1], int(n))
  vec_x.append(sum(y))

plt.scatter(vec_n, vec_x)
plt.show()
```

```
# Chapter 3, problem 19
import numpy as np
import matplotlib.pyplot as plt

vec_n = np.array([1, 5, 25, 50, 100])
vec_var = []
vec_mean = []

for n in vec_n:
    xbar = []
    for i in range(500):
        y = np.random.uniform(0, 1, n)
        xbar.append(np.mean(y)) 
    vec_var.append(np.var(xbar))
    vec_mean.append(np.mean(xbar))

plt.figure()
plt.scatter(vec_n, vec_var)
plt.scatter(vec_n, 1/12./vec_n)
plt.show()

plt.figure()
plt.scatter(vec_n, vec_mean)
plt.plot(vec_n, 0*vec_n + 0.5, 'r')
plt.show()
```
